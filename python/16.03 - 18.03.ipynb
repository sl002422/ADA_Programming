```python
l1 = ['a','c','d','e','f','j','p','b','a']
l2 = ['c','e']


def intersection (l1,l2):
    l3 = [value for value in l1 if value in l2]
    return l3
print(intersection(l1,l2))

intersect = []
for element in l1:
    if element in l2:
        intersect.append(element) #here is the problem that we'll get copies in the results
        #in case there is more than one same element in the ...
        l2.remove(element)
print(intersect)

#another alternative, using sets (if no repeated elements are allowed)
l2 = ['g','c','b','a','m','t','z']
intersect = []
for element in set(l2):
    if element in set(l2):
        intersect.append(element)
print(intersect)
```

```python
#solution to avoid the loop:
l1 = ['a','c','d','e','f','j','p','b','a']
l2 = ['g','c','b','a','m','t','z']
print(list(set(l1) & set(l2)))

#function set can be given an iterable that will be turned to a set without repeating of the variables, so we have set without repeatings as the sets cannot have them (that's the mathematical rule)

print(list(set(l1) | set(l2))) #union
```

```python
paul = ['paul', 1942, True]
mcCartney = paul
print(mcCartney is paul) # they both are contains the same addresses as the mcCartney is not a deep copy
print(id(mcCartney), id(paul)) # the same addresses will be shown
mcCartney = paul.copy() # this is still not the deep copy but at least it copies:)
print(mcCartney is paul)
print(id(mcCartney), id(paul)) # the different addresses will be shown
```
