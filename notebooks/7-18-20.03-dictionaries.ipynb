## Sorting dictionaries
Define a simple function to demonstrate sorting dictionaries.

```python
def sort_dictionary(d, by):
    """Sorting a dictionary by keys or by values.
    - using zip()
    - using operator.itemgetter()
    - using lambda
    """

    # if by == 'k' or by == 'K':
    #     return dict(sorted(zip(d.keys(), d.values())))
    # elif by == 'v' or by == 'V':
    #     return dict(sorted(zip(d.values(), d.keys())))
    # else:
    #     return None

    # from operator import itemgetter
    # if by == 'k' or by == 'K':
    #     return dict(sorted(d.items(), key=itemgetter(0)))
    # elif by == 'v' or by == 'V':
    #     return dict(sorted(d.items(), key=itemgetter(1)))
    # else:
    #     return None

    if by == 'k' or by == 'K':
        return dict(sorted(d.items(), key=lambda item: item[0]))
    elif by == 'v' or by == 'V':
        return dict(sorted(d.items(), key=lambda item: item[1]))
    else:
        return None
```

```python
def demonstrate_dict_sorting():
    """Demonstrate sorting a dictionary.
    """

    songs = {3: 'The Long and Winding Road', 1: 'Oh, Darling', 2: 'The Fool on the Hill'}
    print(sort_dictionary(songs, by='k'))
    print(sort_dictionary(songs, by=12))
```

```python
demonstrate_dict_sorting()
```

## EXERCISE 5
make a game rock-paper-scissors

```python
def RSP ():
    pvp = input('Player 1:') + input('Player 2:')
    print(pvp)
    if pvp == 'pp' or pvp == 'ss' or pvp == 'rr':
        print('draw')
    elif pvp == 'pr' or pvp == 'rs' or pvp == 'sp':
        print('player 1 wins!')
    else:
        print('player 2 wins!')
    replay = input('Replay?\n y/n:')
    if replay == 'y':
        RSP()
RSP()
```

```python
#the same solution but with while loop
def RSP ():
    i = 1
    while i != 0:
        pvp = input('Player 1:') + input('Player 2:')
        print(pvp)
        if pvp == 'pp' or pvp == 'ss' or pvp == 'rr':
            print('draw')
        elif pvp == 'pr' or pvp == 'rs' or pvp == 'sp':
            print('player 1 wins!')
        else:
            print('player 2 wins!')
        replay = input('Replay?\n y/n:')
        if replay == 'n':
            i = 0
RSP()
```

```python
l1 = [1,2,3,4,5,983,4,3,284,3,38,39,43,22]
even = [n for n in l1 if n % 2 == 0]
not_even = [n for n in l1 if n % 2 == 1]
print(f'Even numbers: {even}, not even numbers: {not_even}')
```

```python
#function
l1 = [1,2,3,4,5,983,4,3,284,3,38,39,43,22]

def even (l):
    ln = [n for n in l if n % 2 == 0]
    return ln

print(even(l1))
```
