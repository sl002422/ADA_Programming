```python
s = pd.read_csv('../data/The Beatles songs dataset, v1.csv')
s['Top50'] = 'No'
i = s['Top.50.Billboard'] != -1
s.loc[i, 'Top50'] = 'Yes'
s.loc[s['Top.50.Billboard'] != -1, 'Top50'] = 'Yes'
s.to_csv('../data/The Beatles songs dataset, v2.csv')
s.sample(10)
```

```python
s.loc[i, ['Title','Top50']]
```

```python
s_by_year = s.groupby('Year')
s_by_year.get_group(1963).value_counts('Top50')['Yes']
```

```python
s_by_year.get_group(1963).Top50.value_counts()['Yes']
```

```python
years = np.sort(s.Year.unique())

in_top50 = []
not_in_top50 = []
for y in years:
    g = s_by_year.get_group(y)
    yes = g.Top50.value_counts()['Yes'] if any(g.Top50 == 'Yes') else 0
    no = g.Top50.value_counts()['No'] if any(g.Top50 == 'No') else 0
    in_top50.append(yes)
    not_in_top50.append(no)
display(in_top50)
display(not_in_top50)

df = pd.DataFrame({'In top50': in_top50, 'Not in top50': not_in_top50}, index=years)

df.plot.bar(title='Top50 over the years',
            xlabel='Year',
            ylabel='Count',
            xlim=(1958, 1980),
            ylim=(0,70),
            stacked=True);
```
